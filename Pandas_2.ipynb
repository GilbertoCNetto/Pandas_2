```python
# 1 - Fazer a leitura do arquivo em anexo para um dataframe e mostre o Dataframe todo:
import pandas as pd

dataframe = pd.read_csv('igm_modificado.csv')
display(dataframe)
```

```python
# 2 - Selecione 5 linhas aleatórias do dataframe:

dataframe.sample(5)
```

```python
# 3 - Mostre somente as linhas do índice 3 ao índice 9:

dataframe.iloc[3:10]
```

```python
# 4 - Selecionar as 5 ultimas linhas do dataframe sem usar o comando tail():

ultimhas_5_linhas = dataframe[-5:]
display(ultimhas_5_linhas)
```

```python
# 5 - Mostre 5 linhas aleatórias como uma matriz transposta:

dataframe.sample(5).T
```

```python
# 6 - Mostre a quantidade de cada tipo de porte de cidade existente no dataframe: quantidade de cidades de Pequeno Porte 1, Pequeno Porte 2, Médio Porte e Grande Porte:

dataframe.value_counts('porte')
```

```python
# 7 - Mostre as 5 primeiras linhas usando somente as colunas municipio e indice_governanca:


dataframe[['municipio', 'indice_governanca']].head(5)
```

```python
# 8 - Mostre todas as cidades que são da região Nordeste cujoíndice de governança seja maior que 0.6(somente as colunas município, índice governança, estado e região):
# Filtrar apenas as cidades da região Nordeste
nordeste = dataframe[dataframe['regiao'] == 'NORDESTE']

# Filtrar as cidades com índice de governança maior que 0.6
indice_maior_06 = nordeste[nordeste['indice_governanca'] > 0.6]

# Selecionar apenas as colunas desejadas
resultado = indice_maior_06[['municipio', 'indice_governanca', 'estado', 'regiao']]

display(resultado)
```

```python
# 9 - Mostre o índice de governança agrupado por região:

# Value_Counts do índice de governança agrupado por região
print(dataframe['regiao'].value_counts('indice_governanca'))

# Média do índice de governança agrupado por região
media_indice_por_regiao = dataframe.groupby('regiao')['indice_governanca'].mean()

print('\n', media_indice_por_regiao)

# Soma do índice de governança agrupado por região
soma_indice_por_regiao = dataframe.groupby('regiao')['indice_governanca'].sum()

print('\n', soma_indice_por_regiao)
```

```python
# 10 - Faça um gráfico de barras horizontais com as quantidades de portes de cidades;Resultado esperado:
import matplotlib.pyplot as plt

quantidade_portes = dataframe['porte'].value_counts()

quantidade_portes.plot(kind='barh', color='royalblue')
plt.xlabel('Quantidade')
plt.ylabel('Porte da Cidade')
plt.title('Quantidade de Cidades por Porte')
plt.show()
```

```python
# 11 - Mostre a quantidade de munícipios por estado, classificado em ordem crescente (estado que tem menor quantidade de municípios até o que tem mais municípios):
municipios_por_estado = dataframe.groupby('estado')['municipio'].count()
display(municipios_por_estado.sort_values(ascending = True))
```
